### 종목명

```python
res = req.get('https://finance.naver.com/sise/sise_quant.naver')
res
```

```python
res.text
```

```python
soup = bs(res.text, "html.parser")
```

```python
soup
```

```python
tltle = soup.select('a.tltle')
tltle
```

```python
len(tltle)
```

```python
tltle_list= []
for i in tltle:
    tltle_list.append(i.text)
```

### 현재가

```python
price = soup.select('td.number')
price
```

```python
len(price)
```

```python
price_raw_list = []
for i in price:
    price_raw_list.append(i.text)
```

```python
price_raw_list
```

```python
price_list= []
for i in range(0,1000):
    if i % 10 == 0:
        price = price_raw_list[i]
        price_list.append(price)
```

```python
price_list
```

```python
len(price_list)
```

### 판다스 라이브러리

```python
len(price_list), len(tltle_list)
```

```python
data = {'종목명' : tltle_list, '가격정보' : price_list}
```
