```python
Dataurl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(Dataurl, encoding='euc-kr')
df
```

```python
#quantity 칼럼 값이 3인 데이터를 추출하고, 첫 5행을 출력해주세요
df.loc[df['quantity'] == 3].head()
```

```python
#인덱스 리셋
df.loc[df['quantity'] == 3].head().reset_index(drop=True)
```

```python
# 기존 인덱스를 버지리않고 리셋
df.loc[df['quantity'] == 3].head().reset_index(drop=False)
```

```python
# quantity, item_price 총 2개의 컬럼으로 구성된 데이터 프레임을 만들어주세요
a = df[['quantity','item_price']]
a
```

```python
# 1. item_price의 $표시를 없애주고
# 2. float 타입으로 바꿔주세요
# 3. 그 결과를 new_price라는 컬럼에 저장해주세요
df['new_price'] = df['item_price'].str[1:].astype('float')
a = df['new_price'].head()
a
```

```python
# new price 칼럼이 5이하인 값을 가지는 데이터 프레임을 추출하고,
# 그 갯수를 구해주세요
b = len(df.loc[df.new_price <= 5])
b
```

```python
# item_name이 chicken Salad Bowl인 데이터 프레임을 추출하고,
# index 값을 초기화해주세요
c = df.loc[df.item_name == 'Chicken Salad Bowl'].reset_index(drop=True)
c.head(8)
```

```python
# item_name이 Chicken Salad Bowl이고 new_price 값이 10 이하인 데이터프레임 추출
d = df.loc[(df.item_name == 'Chicken Salad Bowl') | (df.new_price <= 10)]
d.head(3)
```

```python
# new_price 컬럼의 값을 오름차순으로 정리하고
# index를 초기화해주세요
e = df.sort_values('new_price').reset_index(drop=True)
e.tail(3)
```

```python
# df의 item_name에서 메뉴 이름에 'Chips' 가 들어가는 데이터 출력
f = df.loc[df.item_name.str.contains('Chips')]
f.head(6)
```

```python
# df의 new_price 컬럼 값들의 내림차순으로 정리하고
# index를 초기화해주세요
g = df.sort_values('new_price',ascending = False).reset_index(drop=True)
g.tail(3)
```

```python
# df의 item_name 칼럼 값이 Steak Salad 또는 Bowl 인 데이터를 뽑아주세요
h = df.loc[(df.item_name=='Steak Salad') | (df.item_name=='Bowl')]
h
```
