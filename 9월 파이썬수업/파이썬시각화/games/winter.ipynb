```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('winter.csv', encoding='cp949')
df
```

```python
#데이터에서 한국(KOR)데이터만 출력
kr = df[df.Country == 'KOR']
kr.head(3)
```

##### pivot table

```python
# 연도별 메달의 수는? -> 결측치 값을 0으로 채워넣기
# aggfunc = (' ') 안에 원하는 것 넣기(size, mean 같은것)
a = kr.pivot_table(index='Year',columns='Medal',aggfunc='size').fillna('0')
a
```

```python
#pivot table : Sport 별로 성별 수를 구해보기
b = kr.pivot_table(index="Sport", columns='Gender', aggfunc='size')
b
```
