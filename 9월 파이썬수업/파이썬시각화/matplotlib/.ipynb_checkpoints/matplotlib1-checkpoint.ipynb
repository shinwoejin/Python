- matplotlib 모듈(라이브러리)에서 pyplot 을 import 하고
- 앞으로 plt라는 이름으로 부르겠다.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)
```

### Line Plot 그리기

```python
y = [2,4,6]
plt.plot(y)
plt.show() #보여달라는뜻
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
plt.show() #보여달라는뜻
```

### Line Style 설정하기
-  '-',  '--' , '-.' , ':'

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls = '--')
plt.show()
```

### Marker 찍기


```python
'.' 'o' 'v' '^' '<' '>' 's' 'p' '|' '1' '2' '3' '4' '*' 'h' 'H' '+' 'D' '_'
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls = '--', marker = 'o')
plt.show()
```

### Line width

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o', lw = 5)
plt.show()
```

### Line Color

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o', lw = 5 , color = 'r')
plt.show()
```

#### Line Color 종류

'b' = blue 'g' = green 'r' = red 
'c' = cyan 'm' = magenta 'y' = yellow
'k' = black 'w' = white

### Marker color

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o', 
         lw = 5 , color = 'b', markerfacecolor = 'c')
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o', 
         lw = 5 , color = 'k', markerfacecolor = 'c', markersize = 10)
plt.show()
```

### 연습문제

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '--', marker = 'o', 
          c = 'g', mfc = 'g', ms = 7)
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = ':', marker = 'p', 
          c = 'r', mfc = 'b', ms = 7)
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '-.', marker = '^', mec='k'
          ,c = 'm', mfc = 'k', ms = 10)
plt.show()
```

#### 그림 범위 지정

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls='-.',c='m',marker='<',mec='k',mfc='k',ms=10)
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.grid() #뒤에 줄 만들기
plt.show() 
```

### 틱 설정하기

```python
# 틱 설정
# xticks : x축 틱 설정
# yticks : y축 틱 설정
plt.plot(x,y,ls='--',lw=5,marker='o',ms=15
         ,mec='g',mew=5,mfc='r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

### 여러 개 plot 그리기

```python
# 여러개 그래프 그리기
# 한 그림안에 여러개 그리프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x) #그래프 그리기
plt.plot(y)
plt.plot(z)
plt.show() # 위에서 그린 그래프 출력
```

#### 범례 표시(Eng)

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.legend()
plt.show()
```

### 차트에서 한글 보이게 하기

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
plt.legend()
plt.show()
```

### Title 표시

```python
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
plt.legend()
plt.title('범례 표시')
plt.show()
```

### x축, y축 title 표시

```python
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
plt.xlabel('x축')
plt.ylabel('y축').set_rotation(0) #title 돌리기 0 90 180 270
plt.title('범례 표시')
plt.legend()
plt.show()
```
