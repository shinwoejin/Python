```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
from tqdm import tqdm_notebook as tqdm
import requests as req
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 17, 8
rcParams['font.family'] ='AppleGothic'
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.youtube.com/results?search_query=%EA%B0%80%EC%A0%95%EC%9A%A9+%EC%BB%A4%ED%94%BC%EB%A8%B8%EC%8B%A0+&sp=CAM%253D')
```

```python
body = driver.find_element_by_css_selector('body')

while True:
    last = bs(driver.page_source, 'html.parser')
    for i in range(20):
        body.send_keys(Keys.END)
    new = bs(driver.page_source, 'html.parser')
    
    if last == new:
        break
    else:
        pass
```

```python
view = driver.find_elements_by_css_selector('#metadata-line > span:nth-child(1)')
title = driver.find_elements_by_css_selector('#video-title')
```

```python
len(view), len(title)
```

```python
view_list = []
for i in view:
    string = i.text
    string = string.replace('조회수 ', '')
    string = string.replace('회', '')
    string = string.replace('.', '')
    string = string.replace('천', '000')
    string = string.replace('만', '0000')
    string = int(string)
    view_list.append(string)
```

```python
title_list = []
for i in title:
    title_list.append(i.text)
```

```python
data = {'제목' : title_list, '조회수' : view_list}
```

```python
Youtube = pd.DataFrame(data)
Youtube = Youtube.head(59)
```
