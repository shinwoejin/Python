```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
res = req.get('http://music.bugs.co.kr/chart')
res
```

```python
res.text
```

```python
soup = bs(res.text, "html.parser")
```

### 랭킹

```python
ranking = soup.select('div.ranking > strong')
```

```python
len(ranking)
```

```python
ranking_list = []
for i in ranking:
    ranking_list.append(i.text)
```

```python
ranking_list
```

### 타이틀

```python
title = soup.select('p.title > a')
```

```python
title
```

```python
for i in title:
    print(i.text)
```

```python
title_list = []
for i in title:
    title_list.append(i.text)
```

```python
len(title)
```
