### 데이터 전처리

#### 1. 결측치 처리
- 제거
- 채우기

```python
import pandas as pd
import numpy as np
```

```python
d = {'score1' : [100, 90, np.nan, 95],
     'score2' : [30, np.nan, 45, 56],
     'score3' : [52, 40, 80, 98],
     'score4' : [np.nan, np.nan, np.nan, 65]}
```

```python
df = pd.DataFrame(d)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
#null값의 개수 확인
```

```python
df[df.score1.isnull()]
```

#### 행을 기준으로 삭제

```python
df.dropna()
```

```python
df.dropna(axis=0, how='any')
#how='any' : 결측치가 있는 행 전부를 지우겠다
```

#### 열을 기준으로 삭제

```python
df.dropna(axis=1)
```

#### 행의 전체값이 결측치인 행을 삭제

```python
d2 = {'score1' : [100, np.nan, np.nan, 95],
     'score2' : [np.nan, np.nan, np.nan, np.nan],
     'score3' : [52, np.nan, 80, 98],
     'score4' : [np.nan, np.nan, np.nan, 65]}
```

```python
df2 = pd.DataFrame(d2)
df2
```

```python
df2.dropna(how='all')
# how = 'all' 모든값이 NaN일때만 지움
```

#### 임계치 설정해서 제거

```python
df2.dropna(thresh=2)
```

적당한 임계치는 분석가가 판단해서 잘 설정해야 함

#### 특정 열 안에서만 삭제

```python
df2.dropna(subset=['score2','score4'])
```

#### 결측치 채우기

```python
# 특정한 단위값으로 채우고싶을때 사용
df.fillna(5)
```

#### 결측치 바로 이전의 값으로 채우기


```python
df.fillna(method='pad')
# 이전의 값이 비어있다면 채워지지 않고 그대로 NaN을 유지하게됨
```

#### 결측치 바로 뒤의 값으로 채우기

```python
df.fillna(method='bfill')
# 뒤의 값이 비어있으면 NaN 값 유지
```

#### 결측치를 각 열의 평균값으로 채우기

```python
df.fillna(df.mean())
```

#### 결측치를 각 열의 중간값, 최소값, 최대값 채우기

```python
df.fillna(df.median()) #중간값
```

```python
df.fillna(df.min()) #최소값
```

```python
df.fillna(df.max()) #최대값
```

#### replace() 함수로 결측치 채우기
- 지정한 값으로 결측치들이 모두 채워지게 됨(이 부분은 fillna랑 같음
- replace 함수는 nan값 대신에 특정한 문자를 다른무자로 바꾸고 싶을 때 유용하게 사용됨

```python
df.replace(to_replace=np.nan, value=df.median())
```

#### interpolate() 함수로 결측치 채우기

```python
# 앞과 뒤의 행 중간 값으로 채워짐
df.interpolate(method='linear', limit_direction='forward')
```

```python
df.interpolate(method='pad', limit=2)
```

#### 2. 범주형 데이터를 원핫인코딩으로 변경
- 모든 데이터를 0과 1로 변환
- 기계는 모든 데이터를 이진법으로 처리하기 때문에 레이블인코딩보다는 원핫인코딩을 주로 사용한다

```python
df = pd.read_csv('mushrooms.csv')
```

```python
one = pd.get_dummies(df) #원핫인코딩으로 변경
one
```

- get dummies 의 문제점

train 데이터에만 있고 test 데이터에는 없는 카테고리를 원핫인코딩된 컬럼으로 바꿔주지 못함

- sklearn OneHoeEncoder 사용

```python
x = df.iloc[:,1:]
y = df['class']
```

```python
print(len(x))
print(len(y))
```

```python
from sklearn.preprocessing import OneHotEncoder

one = OneHotEncoder(sparse=False) #sparse=False생략 가능
one.fit(x[['cap-shape']])
```

```python
train_cap = one.transform(x[['cap-shape']])
```

```python
train_cap
```

```python
one.categories_
```

```python
o = pd.DataFrame(train_cap, columns=['cap-shape_'
                                     +col for col in one.categories_[0]])
```

```python
one_x = pd.concat([x.drop(columns=['cap-shape']),o], axis=1)
one_x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(one_x,y,
                                                   random_state=0)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

### 3. 범주특성을 레이블 인코딩으로 변환
- 숫자의 크고 작음에 대한 특성이 작용함
- 회귀와 같이 연속된 실수를 다루는 알고리즘에서 레이블 인코딩을 사용하면 잘못된 결과값이 나올 수도 있음

```python
one_x['cap-surface'].unique()
```

```python
#cap-surface의 유니크한 카테고리 데이터 개수 확인
one_x['cap-surface'].value_counts()
```

```python
#인코딩전
one_x['cap-surface']
```

```python
#인코딩 후 - map 사용
one_x['cap-surface'] = one_x['cap-surface'].map({'y':0,'s':1,'f':2,'g':3})
```

```python
one_x['cap-surface'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
encoder.fit(one_x['cap-color']) #학습
```

```python
one_x['cap-color'] = encoder.transform(one_x['cap-color'])
```

```python
one_x
```

```python
one_x['cap-color'].value_counts()
```
