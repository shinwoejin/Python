```python
#target이 명목척도인데 0,1,2로 입력시켜서 서열이나 등간척도로 인식할 수 있어서 
#강제로 명목척도로 인식시켜주기 위해서 권한인코딩을 해서 바꿔줘야함
setosa=[]
versicolor=[]
verginica=[]
for i in df["target"]:
    if i==0:
        setosa.append(1)
        versicolor.append(0)
        verginica.append(0)
    elif i==1:
        setosa.append(0)
        versicolor.append(1)
        verginica.append(0)
    else:
        setosa.append(0)
        versicolor.append(0)
        verginica.append(1)
df["is setosa"]=setosa
df["is versicolor"]=versicolor
df["is verginica"]=verginica
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
prediction = model.predict(X_test)
prediction
```

```python
metrics.accuracy_score(prediction,y_test)
```

```python
from tqdm import tqdm
train_list=[]
test_list=[]

for i in tqdm(range(1,100)):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    train_list.append(model.score(X_train,y_train))
    test_list.append(model.score(X_test,y_test))
    
plt.figure(figsize=(15,5))
plt.plot(range(1,100),train_list,label="train")
plt.plot(range(1,100),test_list,label="test")
plt.legend(loc="upper right")
plt.show()
```
